```
import retro
from stable_baselines.common.policies import MlpPolicy, MlpLstmPolicy, MlpLnLstmPolicy, CnnLnLstmPolicy, CnnPolicy, CnnLstmPolicy
from stable_baselines.common.vec_env import SubprocVecEnv, DummyVecEnv
from stable_baselines import PPO2, A2C
```

```
retro.RetroEnv.
```

```
env = DummyVecEnv([lambda: retro.make('Airstriker-Genesis', state='Level1', record='/gdrive/My Drive/RL/Sonic/')])
```

```
modelname = 'sonicppo'
model = PPO2(CnnPolicy,env,n_steps=256, verbose=1)
model.learn(total_timesteps=10000)
model.save("/gdrive/My Drive/RL/Sonic/" + modelname)
```

```
obs = env.reset()
done = False
reward = 0

while not done:
  actions, _ = model.predict(obs)
  obs, rew, done, info = env.step(actions)
  reward += rew

  
print(rew)
```
